```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB,BernoulliNB,MultinomialNB    # 3 types
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.feature_extraction.text import TfidfVectorizer            #1 vectorize
from sklearn.feature_selection import SelectPercentile                 #2 selector
```

```python
data=pd.read_csv('spam.csv')
```

```python
data
```

```python
x_train,x_test,y_train,y_test=train_test_split(data.Message,data.Category,test_size=0.3,random_state=10)
```

```python
vectorizer=TfidfVectorizer()                      # 1) vecorizer
x_train_transformed=vectorizer.fit_transform(x_train)
x_test_transformed=vectorizer.transform(x_test)
```

```python
feature_names=vectorizer.get_feature_names()
```

```python
feature_names
```

```python
selector=SelectPercentile(percentile=10)                 # 2)selector
selector.fit(x_train_transformed,y_train)
x_train_transformed=selector.transform(x_train_transformed).toarray()
x_test_transformed=selector.transform(x_test_transformed).toarray()
```

```python
model_gaussiannb=GaussianNB()                          #Gaussian
model_gaussiannb.fit(x_train_transformed,y_train)
pred=model_gaussiannb.predict(x_test_transformed)
accuracy_score(y_test,pred)
```

```python
model_bernaulisnb=BernoulliNB()                      #Bernouli
model_bernaulisnb.fit(x_train_transformed,y_train)
pred=model_bernaulisnb.predict(x_test_transformed)
accuracy_score(y_test,pred)
```

```python
model_mulnb=MultinomialNB()                             #Multinomial
model_mulnb.fit(x_train_transformed,y_train)
pred=model_mulnb.predict(x_test_transformed)
accuracy_score(y_test,pred)
```
